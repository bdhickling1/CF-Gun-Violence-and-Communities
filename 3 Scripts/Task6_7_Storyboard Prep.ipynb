## 2. Wrangle data sets

### 2.1 df_HUD (count of assisted housing properties)

```python
# Create count of assisted housing properties per area
assisted_housing = df_HUD.groupby('FIPS Code')['HAS ACTIVE FINANCING'].count().reset_index(name = 'assisted_housing')
assisted_housing
```

```python
# Create a count of properties still receiving financing by area as a new variable
active_finance = df_HUD.groupby('FIPS Code')['HAS ACTIVE FINANCING'].apply(lambda x: (x == 'Y').sum()).reset_index(name='active_assisted_housing')
active_finance
```

```python
# Create a new df_HUD_count data frame with the two new variables
df_HUD_count = assisted_housing.merge(active_finance, on = 'FIPS Code')
```

### 2.3 df_edu to df_edu_dip

```python
# Remove education variables keeping only the % of adults with no high school diploma
df_edu.columns.values.tolist()
```

```python
df_edu_dip = df_edu[['FIPS Code', 'Percent of adults with less than a high school diploma, 1980', 'Percent of adults with less than a high school diploma, 1990', 'Percent of adults with less than a high school diploma, 2000', 'Percent of adults with less than a high school diploma, 2015-19']]
```

```python
# Rename columns
df_edu_dip.rename(columns = {'Percent of adults with less than a high school diploma, 1980': '%adults<high_school_dip_1980', 'Percent of adults with less than a high school diploma, 1990': '%adults<high_school_dip_1990', 'Percent of adults with less than a high school diploma, 2000': '%adults<high_school_dip_2000', 'Percent of adults with less than a high school diploma, 2015-19': '%adults<high_school_dip_2015-19'}, inplace=True)
```

```python
# Create a new variable averaging the % of adults without a high school diploma across all four periods
df_edu_dip['avg_%adults<high_school_dip'] = df_edu_dip[['%adults<high_school_dip_1980', '%adults<high_school_dip_1990', '%adults<high_school_dip_2000', '%adults<high_school_dip_2015-19']].mean(axis=1)
```
