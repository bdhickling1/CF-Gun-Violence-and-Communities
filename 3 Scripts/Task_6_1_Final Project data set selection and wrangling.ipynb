As this process takes a long time to run in Jupyter notebooks, I decided to export the cleaned 'Gun Violence Data' set as a csv file and import it into a different script to carry out the conversion of longitude and latitude values to FIPS Codes. This script is called 'Conversion of Longtitude and Latitude to FIPS Codes and can be found in the 'Scripts' folder of this project.__

__Then, to assist analysis and visualization, I will also use a State, County, and State & County FIPS code mapping tool from the US Unemployment & Education Level data set - which I edited and converted to a .csv file__
